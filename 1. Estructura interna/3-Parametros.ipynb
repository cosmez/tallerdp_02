```python

from google.colab import userdata
import google.generativeai as genai
import os
```

```python
#Accedemos a gemini
genai.configure(api_key=userdata.get('GOOGLE_API_KEY'))

#Revisamos los modelos existentes
for model in genai.list_models():
    print(model.name)
```

Abrimos el CSV de productos que se genero del taller "Web Scrapping"

```python
import pandas as pd
df = pd.read_csv("https://storage.googleapis.com/tallerdp_publico/taller02/products.csv")
```

Obtenemos la fila 1:

```python
product = df.iloc[1]
product_name = product["product_name"]
manufacturer = product["manufacturer"]
```

Generamos el prompt y se lo asignamos a la variable "prompt_template":

```python
prompt_template = f"""

Escribe una descripción publicitaria para el siguiente producto, enfocada en destacar sus beneficios clave de forma clara y atractiva:

Nombre del producto: {product_name}
Marca: {manufacturer}
"""
print(prompt_template)
```

Elegimos el modelo y se lo asignamos a la variable model:

```python
model = genai.GenerativeModel("models/gemini-2.0-flash")
```

Enviamos el prompt, el modelo y los parametros a Gemini:

```python

response = model.generate_content(
            prompt_template,
            generation_config=genai.types.GenerationConfig(
                temperature=1,
                top_p=0.5,
                max_output_tokens=200,
                stop_sequences=["\n"]
            )
        )
print("\n Descripción generada:")
print(response.text.strip())
```

### Multiples parametros

```python
param_sets = [
    {
        "name": "Conservador / Directo",
        "params": {
            "temperature": 0.2,    # Respuesta muy controlada, casi sin creatividad
            "top_p": 0.7,          # Selección más restringida de palabras
            "max_output_tokens": 120,
            "stop_sequences": ["\n"]
        },
        "esperado": "Texto claro, directo, predecible, poco expresivo"
    },
    {
        "name": "Creativo / Libre",
        "params": {
            "temperature": 2.0,    # Muy creativo, puede ser exagerado o impredecible
            "top_p": 1.0,          # Considera todo el vocabulario
            "max_output_tokens": 250,
            "stop_sequences": []
        },
        "esperado": "Texto extenso, imaginativo, muy estilizado"
    },
    {
        "name": "Equilibrado",
        "params": {
            "temperature": 0.7,    # Buen balance entre creatividad y control
            "top_p": 0.9,
            "max_output_tokens": 150,
            "stop_sequences": ["\n"]
        },
        "esperado": "Texto natural, promocional, enfocado pero con estilo"
    },
    {
        "name": "Creativo controlado",
        "params": {
            "temperature": 1.2,    # Bastante creativo pero no tan aleatorio
            "top_p": 0.6,          # Palabras seleccionadas entre las más seguras
            "max_output_tokens": 180,
            "stop_sequences": []
        },
        "esperado": "Texto creativo pero conciso, original con algo de estructura"
    }
]
```

```python
model = genai.GenerativeModel("models/gemini-1.5-flash")
```

```python
for i, setting in enumerate(param_sets, start=1):
    print(f"\n Respuesta {i}: {setting['name']}")
    print(f" Esperado: {setting['esperado']}")
    print(f" Parámetros: {setting['params']}")
    try:
        response = model.generate_content(
            prompt_template,
            generation_config=genai.GenerationConfig(**setting["params"])
        )
        print("Resultado:")
        print(response.text.strip())
        print("--------------------------------------------------------------")
        print("\n")
    except Exception as e:
        print(f" Error al generar contenido: {e}")
```
